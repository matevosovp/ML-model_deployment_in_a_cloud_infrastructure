```python

import joblib
from pathlib import Path

model_path = Path("services/models/Sprint2_cb.pkl")
m = joblib.load(model_path)

print("model type:", type(m))

# Попробуем достать список фичей (если sklearn совместимый объект)
if hasattr(m, "feature_names_in_"):
    print("feature_names_in_:", list(m.feature_names_in_))
else:
    # Если это pipeline, фичи могут быть у отдельных шагов
    if hasattr(m, "named_steps"):
        for name, step in m.named_steps.items():
            if hasattr(step, "feature_names_in_"):
                print(f"step {name} feature_names_in_:", list(step.feature_names_in_))

# Если CatBoost модель внутри, иногда есть feature_names_
if hasattr(m, "feature_names_"):
    print("feature_names_:", list(m.feature_names_))

```
